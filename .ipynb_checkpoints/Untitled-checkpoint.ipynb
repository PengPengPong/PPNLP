```python
def Clean_Stop_Words(JobTitle,StopWords):   # 从职位中去掉停用词
    for stopwords in StopWords:
        if len(stopwords)==1:
            continue
        if len(JobTitle.replace(stopwords,''))>1:
            JobTitle=JobTitle.replace(stopwords,'')
    return JobTitle
```

```python
words=[]
with open('/Users/pp/pycharmprojects/nlp/new_words.txt') as f:
    lines=f.readlines()
    for line in lines:
        words.append(line.split('\t'))

words=map(lambda x:[x[0],float(x[1]),float(x[2]),float(x[3])],words)

words.sort(key=lambda x:(x[1],x[2],x[3]),reverse=True)

cleanwords=[]
for word,freq,ent,cog in words:
    if word.isalpha():
        continue
    if ent>0.5 and cog>100:
        cleanwords.append([word,freq,ent,cog])
        print word,freq,ent,cog


        
print len(cleanwords)
```
